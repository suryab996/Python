```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt

```

```python
# Load in the dataframe
df = pd.read_excel("Enough_Tech.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df['q23'] = df['q23'].astype(str)
```

```python
df.dtypes
```

```python
# Start with one review:
text = df.q23[0]
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
```

```python
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(review for review in df.q23)
#print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stops = set(STOPWORDS)
stops.update(["to","nan","home","able","buy","anything","item","maybe","good","think","work","will","really","new",
              "need","week","going","now","shopping","birthday","go","usually","next","hoping","feel","time","family",
              "looking","make","nice","something","love","items","things","moment","shop","no","Nope","year",
              "nope","Tesco","tesco "," tesco","Tes","TESCO","Tesco ","better","the","the store","needs to","ha",
              "thing","due to","the company","store","staff","doe","to  use"])
              



```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stops ,background_color="white").generate(text)
```

```python
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
